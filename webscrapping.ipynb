```python
import requests
```

```python
# Making a GET request
d=requests.get('https://www.geeksforgeeks.org/python-web-scraping-tutorial/')

# print content of request
print(d.content)
```

```python
from bs4 import BeautifulSoup

# Parsing the HTML
code = BeautifulSoup(d.content, 'html.parser')
print(code.prettify())
```
